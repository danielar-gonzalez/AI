# A1.1 - Aprendizaje estadístico-automático

1- Primero se importó **pandas**, una _librería_ que sirve para trabajar con _tablas de datos_. Luego, se cargó el archivo **"A1.1 Obesidad.csv"** con la función **`read_csv`**, lo que creó una tabla llamada _DataFrame_ con la información del archivo. Por último, se usó el comando **`head(10)`** para ver las _primeras 10 filas de la tabla_ y revisar rápidamente los datos.

```python
import pandas as pd  # carga la librería pandas.
data = pd.read_csv("A1.1 Obesidad.csv")  # lee el archivo CSV.
print(data.head(10))  # muestra las primeras 10 filas del conjunto de datos.
```

2- Una variable que hubiera decidido medir si formara parte del grupo de investigación sería el **nivel de actividad física**, que puede ser una variable **cuantitativa o cualitativa**, dependiendo de cómo se mida. Se podría medir cuantitativamente registrando el número de horas de actividad física realizadas por semana, o cualitativamente, clasificando el nivel de actividad en categorías como "baja", "moderada" o "alta". Otra variable que hubiera decidido medir sería el **consumo de alimentos procesados**, que es una variable **cualitativa**. Una dieta alta en alimentos procesados está relacionada con el aumento de peso y la obesidad. El consumo de estos alimentos podría clasificarse en categorías como "bajo", "moderado" o "alto".


3- En un estudio de inferencia, se utilizaría **`NivelDeObesidad`** como la _variable dependiente_ para analizar cómo las demás variables influyen en los diferentes niveles de obesidad. Esta variable es **cualitativa**, por lo que, en lugar de calcular un promedio como se haría con una variable cuantitativa, se determinaría el _número de registros asociados_ a cada categoría dentro de esa columna mediante la función **`value_counts()`**.

```python
pd.value_counts(data.NivelDeObesidad)  # cuenta las ocurrencias de cada categoría en la columna 'NivelDeObesidad'.
```

4- Se utiliza la librería **`matplotlib`** para crear un _scatter plot_. Primero, se seleccionanaron los datos de la columna **`Edad`** y **`NivelDeObesidad`** del DataFrame **`data`**, filtrando solo las filas donde el sexo es **_femenino_**. Luego, se genera el gráfico de dispersión utilizando los _datos de edad en el eje X_ y los _niveles de obesidad en el eje Y_. El parámetro **`alpha=0.02`** ajusta la transparencia de los puntos. Finalmente, se muestra el gráfico con **`plt.show()`**.

```python
import matplotlib.pyplot as plt  # importa la librería para crear gráficos.
data_x = data.Edad[data.Sexo == "Female"]  # filtra los datos de edad para mujeres.
data_y = data.NivelDeObesidad[data.Sexo == "Female"]  # filtra los datos de nivel de obesidad para mujeres.
plt.figure(figsize=(8,4))  # crea una figura de 8x4.
plt.scatter(x=data_x, y=data_y, alpha=0.02)  # crea un gráfico con los datos filtrados.
plt.show()  # muestra el gráfico.
```
